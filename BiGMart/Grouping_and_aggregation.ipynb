```python
df2 = df.rename({'First Name':'First_Name','Start Date':'Start_Date', 'Last Login Time':'Last_Login_Time',
       'Bonus %':'Bonus%', 'Senior Management':'Senior_Management'},axis=1)
```

```python
cont_df2 = df2.select_dtypes(exclude='object')
cont_df2
```

```python
### Plot to view distribution
def plot_histplot(df2):
    plt.figure(figsize=(20,20))
    for i,col in enumerate(df2.columns):
        plt.subplot(2,2,i+1)
        plt.hist(df2[col])
        plt.title(col)
    plt.tight_layout()
    plt.show()
```

```python
plot_histplot(cont_df2)
```

```python
def plot_boxtplot(df2):
    plt.figure(figsize=(20,20))
    for i,col in enumerate(df2.columns):
        plt.subplot(2,2,i+1)
        plt.boxplot(df2[col])
        plt.title(col)
    plt.tight_layout()
    plt.show()
```

```python
plot_boxtplot(cont_df2)
```
